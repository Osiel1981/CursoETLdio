```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['*****', '***', '****', '###!', '####', 'NULL', '**']
df_inicial = pd.read_csv('ocorrencia.csv', delimiter=';', parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df_inicial.head(10)
```

```python
df_inicial.dtypes
```

```python
df_inicial = df_inicial.loc[:,['codigo_ocorrencia', 'codigo_ocorrencia2', 'ocorrencia_classificacao', 'ocorrencia_cidade', 'ocorrencia_uf', 'ocorrencia_aerodromo', 'ocorrencia_dia' , 'ocorrencia_hora', 'total_recomendacoes']]
```

```python
schema = pa.DataFrameSchema(
columns = {
    "codigo_ocorrencia":pa.Column(pa.Int, nullable=True),  
    "codigo_ocorrencia2":pa.Column(pa.Int, nullable=True), 
    "ocorrencia_classificacao":pa.Column(pa.String, nullable=True),
    "ocorrencia_cidade":pa.Column(pa.String, nullable=True),
    "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
    "ocorrencia_aerodromo":pa.Column(pa.String, nullable=True),
    "ocorrencia_dia":pa.Column(pa.DateTime, nullable=True),
    "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
    "total_recomendacoes":pa.Column(pa.Int, nullable=True)    
}
)
```

```python
filtro_classificacao1 = (df_inicial.ocorrencia_classificacao == 'INCIDENTE GRAVE')|(df_inicial.ocorrencia_classificacao == 'INCIDENTE')
filtro_classificacao2 = df_inicial.ocorrencia_uf == 'SP'
df_inicial.loc[filtro_classificacao1 & filtro_classificacao2]
```

```python
filtro_classificacao1 = df_inicial.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE'])
filtro_classificacao2 = df_inicial.ocorrencia_uf == 'SP'
df_inicial.loc[filtro_classificacao1 & filtro_classificacao2]
```

```python
filtro_cidade = df_inicial.ocorrencia_cidade.str.contains('TRA|LHA|RIO|CEU')
df_inicial.loc[filtro_cidade]
```

```python
filtro_cidade = df_inicial.ocorrencia_cidade.str[:3] == 'MAR'
df_inicial.loc[filtro_cidade]
```

```python
filtro_dia = df_inicial.ocorrencia_dia.dt.year == 2015
df_inicial.loc[filtro_dia]
```

```python
filtro_dia_mes = (df_inicial.ocorrencia_dia.dt.month == 12) & (df_inicial.ocorrencia_dia.dt.year == 2015) & (df_inicial.ocorrencia_dia.dt.day == 12)
df_inicial.loc[filtro_dia_mes]
```

```python
df_inicial['ocorrencia_data_hora'] = pd.to_datetime(df_inicial.ocorrencia_dia.astype(str) + ' ' + df_inicial.ocorrencia_hora)
df_inicial.ocorrencia_data_hora.astype(str) + ' ' + df_inicial.ocorrencia_hora
```

### Agrupamento de ocorrencias

```python
filtro_data_inicio = df_inicial.ocorrencia_data_hora >= '2015-03-01 00:00:01'
filtro_data_fim = df_inicial.ocorrencia_data_hora <= '2015-03-31 23:59:59'
ocorrencias_03_2015 = df_inicial.loc[filtro_data_inicio & filtro_data_fim]
```
